```python
import numpy as np
import scipy
import matplotlib.pyplot as plt
%matplotlib inline
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)
```

# Binomial : Discrete 

```python
# Per experiment, the machine pokes 100 times
n = 100
# The probability of success is 0.5
p = 0.5
# Assume that we repeat the event for 5000 times
nPoints = 5000
# We can randomly draw sammples from a binomial distribution
xBinom = np.random.binomial(n, p, size=nPoints)
print(xBinom[:20])

# Plot Distribution
plt.figure(figsize=(12,4))                                   # figure size

plt.subplot(121)                                             # subplot 1 (left)
plt.plot(xBinom, "bo", color="purple", alpha=0.5)            # distribution plot, bo meanas using blue circle
plt.xlabel("Number times")
plt.ylabel("No of Success")
plt.grid()

# histogram (subplot 2)
plt.subplot(122)
plt.hist(xBinom, bins=25, color="purple", alpha=0.5)
plt.xlabel("No of Success")
plt.ylabel("Frequency of Number of Success")
```

# Poisson Distribution

```python
# E.g. We want to know how many people walk through a door per hour
# Parameters to tweak(lambdd, ld): the mean occurrence of the events per interval

# Set lambda to 5
ld = 5                                                 # mean No. of people walk through the door per hour
# Set the number of points to 5000 
nPoints = 5000                                         # repeat the experiment for 5000 times

# We can randomly draw sammples from a Poisson distribution
rPoisson=np.random.poisson(ld, size=nPoints)
rPoisson[:20]

# Plot Distribution
plt.figure(2,figsize=(12,4))                           # figure size

plt.subplot(121)                                       # left subplot
plt.plot(rPoisson, "b.", color="purple", alpha=0.5)    # distribution plot, b. means using blue dots
plt.xlabel("No. of Samples")
plt.ylabel("No. of People")

plt.subplot(122)                                       # right subplot
plt.hist(rPoisson, bins=50, color="purple",alpha=0.5)  # plot histogram
plt.xlabel("Number of People")
plt.ylabel("Frequency")
plt.grid()
```

# Normal or Gaussian Distribution

```python
# For normal or Gaussian distribution
# Parameter to tweak: mu and sigma
mu = 5
sigma = 3.0
nPoints = 5000


# We can randomly draw sammples from a normal distribution
rGauss= np.random.normal(mu, sigma, nPoints)

# Plot Distribution
plt.figure(2, figsize=(12,4))

plt.subplot(121)
plt.plot(rGauss, "b.", color="purple", alpha=0.5)

plt.subplot(122)
plt.hist(rGauss, bins=50, color="purple", alpha=0.5)

plt.show()
```

# Binomial Distribution

# Changing the n parameter
When changing the n parameter to 10, 100, and 1000 
the 3 histograms below will show a bell curve shape and as n increases, the frequency of the number of successes is less

```python
# Per experiment, the machine pokes 100 times
n = [10,100,1000]

# The probability of success is 0.5
p = 0.5

# Assume that we repeat the event for 5000 times
nPoints = 5000

# We can randomly draw sammples from a binomial distribution
xBinom = np.random.binomial(n[0], p, size=nPoints)
xBinom1 = np.random.binomial(n[1], p, size=nPoints)
xBinom2 = np.random.binomial(n[2], p, size=nPoints)

print(xBinom[:20])
print(xBinom1[:20])
print(xBinom2[:20])

# Plot Distribution
plt.figure(figsize=(20,13))                                   # figure size

plt.subplot(331)                                             # subplot 1 (left)
plt.hist(xBinom, bins=25, color="purple", alpha=0.5)         # plot histogram

plt.xlabel("No. of Success where n is 10")
plt.ylabel("Frequency of Number of Success")

plt.subplot(332)                                             # subplot 2 (right)
plt.hist(xBinom1, bins=25, color="purple", alpha=0.5)         # plot histogram

plt.xlabel("No. of Success where n is 100")
plt.ylabel("Frequency of Number of Success")

plt.subplot(333)                                             # subplot 2 (right)
plt.hist(xBinom2, bins=25, color="purple", alpha=0.5)         # plot histogram

plt.xlabel("No. of Success where n is 1000")
plt.ylabel("Frequency of Number of Success")
```
